```python
import matplotlib.pyplot as plt

# Sample data
subjects = ['Math', 'Science', 'English', 'History', 'Art']
marks_1 = [12, 45, 56, 76, 67]
colors = ['skyblue', 'lightgreen', 'lightcoral', 'gold', 'plum']

# Create donut chart
plt.figure(figsize=(8, 8))
plt.pie(marks_1, labels=subjects, colors=colors, autopct='%1.1f%%', startangle=90, wedgeprops={'width': 0.4})

plt.title("Donut Chart of Marks (Test 1)")
plt.axis('equal')  # Keep the pie circular
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

subjects = ['Math', 'Science', 'English', 'History', 'Art']
marks_1 = [12, 45, 56, 76, 67]
marks_2 = [23, 56, 78, 54, 34]

# Append first element to close the circle
subjects += [subjects[0]]
marks_1 += [marks_1[0]]
marks_2 += [marks_2[0]]

# Create angles for each axis (same length as data)
angles = np.linspace(0, 2 * np.pi, len(subjects), endpoint=False).tolist()

plt.figure(figsize=(8, 8))
ax = plt.subplot(111, polar=True)

ax.plot(angles, marks_1, label='Test 1', linewidth=2)
ax.fill(angles, marks_1, alpha=0.25)

ax.plot(angles, marks_2, label='Test 2', linewidth=2)
ax.fill(angles, marks_2, alpha=0.25)

ax.set_xticks(angles[:-1])  # set labels except the last (duplicate)
ax.set_xticklabels(subjects[:-1])

plt.title("Radar Chart: Test 1 vs Test 2")
plt.legend(loc='upper right', bbox_to_anchor=(1.1, 1.1))
plt.show()
```
